```python
import pandas as pd
from pathlib import Path
import yfinance as yf
```

```python
amzn = yf.Ticker('amzn')
amzn_df = amzn.history()
amzn_df.head()
```

```python
amzn_df = amzn.history(interval='1d', start='2010-06-01', end='2022-09-26')
amzn_df.head()
```

```python
amzn_df.drop(columns=["Open","High","Low","Volume","Dividends","Stock Splits"], inplace=True)
amzn_df.head()
```

```python
amzn_df.to_csv('amzn.csv')
```
